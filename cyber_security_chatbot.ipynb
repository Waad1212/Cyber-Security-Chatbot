```python

document=[]
for file in os.listdir("docs"):
  if file.endswith(".pdf"):
    pdf_path="./docs/"+file
    loader=PyPDFLoader(pdf_path)
    document.extend(loader.load())
  elif file.endswith('.docx') or file.endswith('.doc'):
    doc_path="./docs/"+file
    loader=Docx2txtLoader(doc_path)
    document.extend(loader.load())
  elif file.endswith('.txt'):
    text_path="./docs/"+file
    loader=TextLoader(text_path)
    document.extend(loader.load())
```

```python
document_splitter=CharacterTextSplitter(separator='/n', chunk_size=500, chunk_overlap=100)
```

```python
document_chunks=document_splitter.split_documents(document)
```

```python
#Set up HuggingFace Pipeline with Llama-2-7b-chat-hf model
model = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model)
pipe=pipeline = transformers.pipeline(
      "text-generation", #task
      model=model,
      tokenizer=tokenizer,
      torch_dtype=torch.bfloat16,
      trust_remote_code=True,
      device_map="auto",
      max_length=1000,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id
)
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm=ChatOpenAI(temperature=0.7, model_name='gpt-3.5-turbo')
```

```python
llm
```

```python
memory=ConversationBufferMemory(memory_key='chat_history', return_messages=True)
```

```python
#Create our Q/A Chain
pdf_qa=ConversationalRetrievalChain.from_llm(llm=llm,retriever=vectordb.as_retriever(search_kwargs={'k':6}),verbose=False, memory=memory)
```

```python


def pdf_qa_interface(query):
    if query == "exit" or query == "quit" or query == "q" or query == "f":
        print('Exiting')
        sys.exit()
    if query == '':
        return ''
    result = pdf_qa({"question": query})
    return result["answer"]

iface = gr.Interface(fn=pdf_qa_interface, inputs="text", outputs="text", title="cyber security chatbot. ask me !")
iface.launch()
```
